# Q-Q PLOT


```python
# 
import numpy as np
import pylab
import scipy.stats as stats
```

```python
# N(0,1)
std_normal = np.random.normal(loc = 0,scale = 1,size = 1000)# loc is computed as mean and scale is defined as spread or variance
# computing 0  to 100th percentile of std-normal
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
#generate 1000 samples from N(20,5)
measurements = np.random.normal(loc = 20,scale = 5,size = 100)
# plotting q-q plot between measurements and norm
stats.probplot(measurements,dist = "norm",plot = pylab)
pylab.show()
```

# OBSERVATION:-

1. #values of x and y lie on same line  then we can conclude that x and y follow the same distribution.
2.if   X~N(0,1) then we can conclude that y also follows gausssian distribution
3. through Q-Q plot we can compute the rough mean value of odered random variable in this Y, u = 19-20.

```python
#generate 50 samples from N(20,5)
measurements = np.random.normal(loc = 20,scale = 5,size = 100)
# plotting q-q plot between measurements and norm
stats.probplot(measurements,dist = "norm",plot = pylab)
pylab.show()
```

```python
#generate 50000 samples from N(20,5)
measurements = np.random.normal(loc = 20,scale = 5,size = 50000)
# plotting q-q plot between measurements and norm
stats.probplot(measurements,dist = "norm",plot = pylab)
pylab.show()
```

```python
# generate 100 samples from N(20,5)
measurements  = np.random.uniform(low = -1,high=1,size = 100)
stats.probplot(measurements,dist = "norm",plot = pylab)
pylab.show()
```
